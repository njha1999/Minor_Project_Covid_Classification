```
import cv2
import os
import numpy as np
import pandas as pd
```

```

img_array = []
base = "temp/data_upload_v2/train/"
for img in os.listdir(base + "covid"):
    img_arr = cv2.imread(base + "covid/"+img, 0)
    img_arr = cv2.resize(img_arr, (128,128))
    img_arr = img_arr.flatten()
    img_arr = np.append(img_arr, 1)
    img_array.append(img_arr)
print(img_array)
```

```
print(len(img_array))
```

```
base = "temp/data_upload_v2/train/"
for img in os.listdir(base + "non"):
    img_arr = cv2.imread(base + "non/"+img, 0)
    img_arr = cv2.resize(img_arr, (128,128))
    img_arr = img_arr.flatten()
    img_arr = np.append(img_arr, 0)
    img_array.append(img_arr)
print(img_array)
```

```
print(len(img_array))
```

```
img_array_test = []
base2 = "temp/data_upload_v2/test/"
for img in os.listdir(base2 + "covid"):
    img_arr = cv2.imread(base2 + "covid/"+img, 0)
    img_arr = cv2.resize(img_arr, (128,128))
    img_arr = img_arr.flatten()
    img_arr = np.append(img_arr, 1)
    img_array_test.append(img_arr)
s = 0
for folder in os.listdir(base2 + "non"):
    for img in os.listdir(base2 + "non/" + folder):
        s+=1
        img_arr = cv2.imread(base2 + "non/"+folder+'/'+img, 0)
        img_arr = cv2.resize(img_arr, (128,128))
        img_arr = img_arr.flatten()
        img_arr = np.append(img_arr, 0)
        img_array_test.append(img_arr)

print(s)
print(len(img_array_test))
```

```
img_array = np.array(img_array)
img_array_test = np.array(img_array_test)
np.random.shuffle(img_array)
np.random.shuffle(img_array_test)
```

```
from sklearn.svm import SVC
clf = SVC()
clf.fit(img_array_test[:, :-1], img_array_test[:, -1])
```

```
print(img_array)
```

```
pred = clf.predict(img_array[:, :-1])
print(np.mean(pred == img_array[:,-1]))
```

```
from sklearn.metrics import confusion_matrix
print(confusion_matrix(img_array[:,-1], pred))
```
